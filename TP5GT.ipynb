```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np 
```

```python
G_X = nx.Graph()
node = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
edge = [('A', 'E'), ('A', 'F'), ('A', 'G'), 
        ('B', 'A'), ('B', 'C'), ('B', 'D'), ('B', 'G'),
        ('C', 'B'), ('C', 'D'), 
        ('D', 'B'), ('D', 'C'), ('D', 'E'), ('D', 'G'),
        ('E', 'A'), ('E', 'D'), ('E', 'F'),
        ('F', 'A'), ('F', 'E'),
        ('G', 'A'), ('G', 'B'), ('G', 'D'), ('G', 'E')]
G_X.add_nodes_from(node)
G_X.add_edges_from(edge)
pos = {'A': (0, 2), 'B': (4, 2), 'C': (5, 1), 'D': (4, 0),
    'E': (0, 0), 'F': (-1, 1), 'G': (2, 1)}

nx.draw_networkx_nodes(G_X, pos, node_color='lightpink', node_size=500,)
nx.draw_networkx_edges(G_X, pos, edge_color='black', width=2)
nx.draw_networkx_labels(G_X, pos, font_size=12, font_family='sans-serif')
plt.title("Undirected Graph X ")
```

```python
G_Y = nx.Graph()
node = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
edge = [('A', 'E'), ('A', 'F'),
        ('B', 'A'), ('B', 'C'), ('B', 'D'), ('B', 'G'),
        ('C', 'B'), ('C', 'D'), 
        ('D', 'B'), ('D', 'C'),
        ('E', 'A'), ('E', 'D'), ('E', 'F'),
        ('F', 'A'), ('F', 'E'),
        ('G', 'B'), ('G', 'E')]
G_Y.add_nodes_from(node)
G_Y.add_edges_from(edge)
pos = {'A': (0, 2), 'B': (4, 2), 'C': (5, 1), 'D': (4, 0),
    'E': (0, 0), 'F': (-1, 1), 'G': (2, 1)}

nx.draw_networkx_nodes(G_Y, pos, node_color='lightpink', node_size=500,)
nx.draw_networkx_edges(G_Y, pos, edge_color='black', width=2)
nx.draw_networkx_labels(G_Y, pos, font_size=12, font_family='sans-serif')
plt.title("Undirected Graph Y ")

```

```python
#1.1 determine Graph X and Graph Y are Hamiltonian or not
print("Graph X is Hamiltonian:", nx.is_hamiltonian(G_X))
print("Graph Y is Hamiltonian:", nx.is_hamiltonian(G_Y))
```
